```python
regions = {
    "Central": """Charlottesville
Colonial Heights
Hopewell
Petersburg
Richmond""".splitlines(),
    "West Central":"""Lynchburg
Radford
Roanoke
Salem""".splitlines(),
    "Southside": """Danville
Emporia
Franklin
Martinsville""".splitlines(),
    "Hampton Roads":"""Chesapeake
Hampton
Newport News
Norfolk
Poquoson
Portsmouth
Suffolk
Virginia Beach
Williamsburg""".splitlines(),
    "Eastern":"""Accomack
Essex
King & Queen
Lancaster
Middlesex
Northhampton
Northumberland
Richmond
Westmoreland""".splitlines(),
    "Southwest":"""Bristol
Galax
Norton""".splitlines(),
    "Northern":"""Alexandria
Fairfax
Falls Church
Fredericksburg
Manassas
Manassas Park""".splitlines(),
    "Southern":"""Buena Vista
Covington
Harrisonburg
Lexington
Stauton
Waynesboro
Winchester""".splitlines()}
```

```python
len(regions)
```

```python
city_to_region = [(city,region) for region, cities in regions.items() for city in cities]
```

```python
city_to_region = pd.DataFrame(city_to_region, columns=["City", "Region"])
city_to_region["City"] = city_to_region["City"].str.upper()
city_to_region
```

```python
merged2 = pd.merge(merged, city_to_region, left_on="employercity", right_on="City")
```

```python
merged2.head()
```

```python
merged2.to_csv("lajc_clean_categorization_region.csv")
```
